```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv("Iris2.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = pd.get_dummies(df, columns=["Species"])
```

```python
x
```

```python
x.to_csv('output.csv', encoding = 'utf-8-sig') 
files.download('output.csv')
```

```python
y = x['Species_Iris-virginica']
```

```python
y
```

```python
x = x.drop(['Id','Species_Iris-virginica','Species_Iris-versicolor'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.30, random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
len(ytest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(xtrain, ytrain)
```

```python
predicted = regression.predict(xtest)
```

```python
predicted
```

```python
regression.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest,predicted)*100
```
